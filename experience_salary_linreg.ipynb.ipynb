Dataset link : https://www.kaggle.com/datasets/saquib7hussain/experience-salary-dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("data/salary_data.csv")
df.head()
```

```python
df = df.rename(
    columns={"exp(in months)": "Experience", "salary(in thousands)": "Salary"}
)
```

```python
X = df[["Experience"]]
y = df["Salary"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=99
)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
print(model.intercept_, model.coef_[0])
```

```python
import pandas as pd

print(
    "Salary for 5 years of experience:",
    model.predict(pd.DataFrame([[5]], columns=["Experience"]))[0],
)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score

print("R2:", r2_score(y_test, y_pred))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.regplot(x=X, y=y, ci=None, line_kws={'color':'#fd8d3c'}, scatter_kws={'color':'#6baed6'})
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Best Fit Line — Salary Prediction")
plt.show()
```
