Dataset link : https://www.kaggle.com/datasets/yasserh/housing-prices-dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("data/house_prices.csv", usecols=["price", "area"])
df.head()
```

```python
X = df[["area"]]
y = df["price"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=99
)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
print(model.intercept_, model.coef_[0])
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score

print("R2:", r2_score(y_test, y_pred))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("white")
sns.regplot(x=X, y=y, ci=None, line_kws={'color':'#fd8d3c'}, scatter_kws={'color':'#6baed6'})
plt.xlabel("House Area (sq ft)")
plt.ylabel("House Price ($)")
plt.title("Regression Plot — House Price Prediction")
plt.show()
```
