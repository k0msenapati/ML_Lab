Dataset link : https://www.kaggle.com/datasets/himanshunakrani/student-study-hours?select=score.csv

```python
import pandas as pd
```

```python
df = pd.read_csv("data/studyhour_score.csv")
df.head()
```

```python
corr = df['Hours'].corr(df['Scores'])
print(f"Correlation between Hours and Scores: {corr:.2f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("white")
sns.scatterplot(x='Hours', y='Scores', data=df, color='#2b8cbe', alpha=0.8)
plt.title("Study Hours vs Scores - Relationship Analysis")
plt.show()
```

```python
X = df[["Hours"]]
y = df["Scores"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=99
)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
print(model.intercept_, model.coef_[0])
```

```python
import pandas as pd

print(
    "Score for 8 hours of study:",
    model.predict(pd.DataFrame([[8]], columns=["Hours"]))[0],
)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score

print("R2:", r2_score(y_test, y_pred))
```

```python
sns.regplot(x=X, y=y, ci=None, line_kws={'color':'#fd8d3c'}, scatter_kws={'color':'#6baed6'})
plt.xlabel("Study Hours")
plt.ylabel("Scores")
plt.title("Regression Plot — Score Prediction")
plt.show()
```
